```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('experiments.csv')
```

```python
df
```

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize = (11,7))
ax = sns.lineplot(x=df['client_threads'], y=df['avg_response'], hue=df['pod_count'])
ax.set(xlabel ="client threads", ylabel = "response time (s)", title ='Relationship between client threads, pods and response time.',
           )
plt.show()
fig.savefig('results_graph.png')
```
